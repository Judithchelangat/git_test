```python
x_str = input("Enter value for x: ")
y_str = input("Enter value for y: ")

try:
    x = float(x_str)
    y = float(y_str)
    result = x + y
    print("The sum is:", result)
except ValueError:
    print("Invalid input. Please enter numbers only.")
```
